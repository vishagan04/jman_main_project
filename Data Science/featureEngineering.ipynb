```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, chi2

# Load the new DataFrame from CSV
data = pd.read_csv('new_dataframe.csv')

# Encode any remaining categorical features if necessary
label_encoder = LabelEncoder()

# Encode 'approvalStatus' if it's still categorical
data['approvalStatus_encoded'] = label_encoder.fit_transform(data['approvalStatus'])

# 1. Correlation Matrix for Numerical Features
# Extract only numerical columns for correlation matrix
numerical_features = ['rating', 'marks', 'role_encoded', 'department_encoded', 
                      'competencyLevel_encoded', 'skill_name_encoded', 
                      'course_name_encoded']

# Compute correlation matrix
correlation_matrix = data[numerical_features].corr()

# Plot the heatmap for correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# 2. Chi-Squared Test for Categorical Features
X = data[['role_encoded', 'department_encoded', 'competencyLevel_encoded', 
          'skill_name_encoded', 'course_name_encoded']]  # Categorical features
y = data['approvalStatus_encoded']  # Target variable (encoded)

# Apply SelectKBest to extract top categorical features based on Chi-Squared test
chi2_selector = SelectKBest(chi2, k='all')
X_kbest = chi2_selector.fit_transform(X, y)

# Display Chi-squared scores
chi2_scores = chi2_selector.scores_
feature_scores = pd.DataFrame({'Feature': X.columns, 'Chi2 Score': chi2_scores})
feature_scores = feature_scores.sort_values(by='Chi2 Score', ascending=False)
print("Chi-Squared Scores for Categorical Features:\n", feature_scores)

# 3. Feature Importance with Random Forest
# Features and target variable for Random Forest
X_rf = data[['role_encoded', 'department_encoded', 'competencyLevel_encoded', 
             'skill_name_encoded', 'course_name_encoded', 'rating', 'marks']]
y_rf = data['approvalStatus_encoded']  # Target variable

# Fit RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_rf, y_rf)

# Get feature importances
feature_importances = rf_model.feature_importances_

# Create a DataFrame for feature importances
feature_importance_df = pd.DataFrame({'Feature': X_rf.columns, 'Importance': feature_importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Display the feature importances
print("Feature Importances from Random Forest:\n", feature_importance_df)

# Plot the feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance Based on Random Forest')
plt.show()
```

```python
import pandas as pd

# Load your data (assuming 'new_dataframe.csv' is structured with the necessary columns)
data = pd.read_csv('new_dataframe.csv')

# Select relevant features for the content-based recommendation
content_data = data[['course_id', 'course_name', 'competencyLevel', 'skill_name']]

# Create a mapping of employee skills
employee_skills = data.groupby('employee_id')['skill_name'].apply(list).reset_index()

def get_content_based_recommendations(employee_id, employee_skills_df, content_data):
    # Get the skills of the employee
    employee_skills_list = employee_skills_df.loc[employee_skills_df['employee_id'] == employee_id, 'skill_name'].values[0]
    
    # Make sure to create a flat set of skills
    employee_skill_set = set(employee_skills_list)

    # Recommend courses that match the employee's skills
    recommended_courses = content_data[content_data['skill_name'].isin(employee_skill_set)]

    # If no courses found, recommend courses that enhance the most common skill
    if recommended_courses.empty:
        most_common_skill = pd.Series(employee_skills_list).value_counts().idxmax()
        recommended_courses = content_data[content_data['skill_name'] == most_common_skill]

    return recommended_courses[['course_id', 'course_name', 'competencyLevel']].drop_duplicates()

# Example usage
print("Content-Based Recommendations for Employee 123:")
recommended_courses = get_content_based_recommendations(123, employee_skills, content_data)
print(recommended_courses)
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# Load your data
data = pd.read_csv('new_dataframe.csv')

# Create user-item matrix for collaborative filtering
collab_data = data.pivot_table(index='employee_id', columns='course_id', values='rating').fillna(0)

# Initialize the KNN model
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(collab_data)

def get_collab_recommendations(employee_id, n_recommendations=5):
    try:
        idx = collab_data.index.get_loc(employee_id)
    except KeyError:
        print(f"Employee ID {employee_id} not found in the dataset.")
        return []

    distances, indices = knn.kneighbors(collab_data.iloc[idx, :].values.reshape(1, -1), n_neighbors=n_recommendations + 1)

    # Get the course recommendations based on the nearest neighbors
    recommended_courses = set()
    for i in indices.flatten()[1:]:  # Skip the first index as it's the employee themselves
        # Use boolean indexing to get the courses with ratings greater than 0
        non_zero_courses = collab_data.columns[collab_data.iloc[i] > 0]
        recommended_courses.update(non_zero_courses.tolist())
    
    return list(recommended_courses)[:n_recommendations]

# Example usage
print("Collaborative Filtering Recommendations for Employee 123:")
recommended_courses = get_collab_recommendations(123, n_recommendations=5)
print(recommended_courses)
```

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
from sklearn.neighbors import NearestNeighbors

# Load your data and filter where approvalStatus is "Approved"
data = pd.read_csv('new_dataframe.csv')  # Ensure this path is correct
filtered_data = data[data['approvalStatus'] == 'Approved']

# Sample 5000 random entries from the filtered data
filtered_data = filtered_data.sample(n=5000, random_state=42)  # Set random_state for reproducibility

# Check for necessary columns in the filtered_data
required_columns = ['employee_id', 'course_id', 'rating']
if not all(col in filtered_data.columns for col in required_columns):
    print(f"Error: The filtered_data must contain the following columns: {required_columns}")
else:
    # Create user-item matrix
    collab_data = filtered_data.pivot_table(index='employee_id', columns='course_id', values='rating').fillna(0)

    # Initialize the KNN model
    knn = NearestNeighbors(metric='cosine', algorithm='brute')
    knn.fit(collab_data)

    def get_collab_recommendations(employee_id, n_recommendations=5):
        """Get course recommendations based on collaborative filtering."""
        try:
            idx = collab_data.index.get_loc(employee_id)
        except KeyError:
            print(f"Employee ID {employee_id} not found in the dataset.")
            return []

        distances, indices = knn.kneighbors(collab_data.iloc[idx, :].values.reshape(1, -1), n_neighbors=n_recommendations + 1)

        # Get the course recommendations based on the nearest neighbors
        recommended_courses = set()
        for i in indices.flatten()[1:]:  # Skip the first index as it's the employee themselves
            non_zero_courses = collab_data.columns[collab_data.iloc[i] > 0]
            recommended_courses.update(non_zero_courses.tolist())

        return list(recommended_courses)[:n_recommendations]

    def evaluate_collab_model(filtered_data, n_recommendations=5, use_binary=True):
        """Evaluate the collaborative filtering model's performance."""
        true_ratings = []
        predicted_ratings = []

        for _, row in filtered_data.iterrows():
            employee_id = row['employee_id']
            course_id = row['course_id']
            true_rating = row['rating']

            # Get recommendations for the employee
            recommended_courses = get_collab_recommendations(employee_id, n_recommendations=n_recommendations)

            # Check if the course is in the recommended courses
            predicted_rating = 1 if (course_id in recommended_courses) else 0
            if not use_binary:  # If using actual ratings
                predicted_rating = true_rating if course_id in recommended_courses else 0
            
            predicted_ratings.append(predicted_rating)
            true_ratings.append(true_rating)

        # Calculate MAE and RMSE
        mae = mean_absolute_error(true_ratings, predicted_ratings)
        rmse = np.sqrt(mean_squared_error(true_ratings, predicted_ratings))



    # Print evaluation metrics
    print(f'Mean Absolute Error (MAE): {mae:.4f}')
    print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')

```
